# Pandas
## Useful functions

```python
# import necessary libraries
import numpy as np
import pandas as pd
```

```python
# Load the data
df = pd.read_csv("./data/tips.csv")
df.head()
```

```python
# Using apply method to single column
def last_four_digits(num):
    return str(num)[-4:]
```

```python
df["cc_last_four"] = df["CC Number"].apply(last_four_digits)
```

```python
df.head()
```

```python
def praise_tip(tip):
    if tip <= 5:
        return "Good"
    elif tip > 5 and tip <= 15:
        return "Better"
    else:
        return "Excellent" 
```

```python
df["tip_praise"] = df["tip"].apply(praise_tip)
df.head()
```

```python
# function to calculate quality of the tip
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Not Generous"
```

```python
# apply the function to the dataframe
df["tip_quality"] = df.apply(lambda df: quality(df["total_bill"], df["tip"]), axis=1)
```

```python
df.head()
```

```python
# achieve above apply using np.vectorize
df["quality_tip"] = np.vectorize(quality)(df["total_bill"], df["tip"])
```
