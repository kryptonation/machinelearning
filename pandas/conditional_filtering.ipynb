## Pandas
### Conditional filtering

```python
# import necessary libraries
import numpy as np
import pandas as pd
```

```python
# Load the data
df = pd.read_csv("./data/tips.csv")
df.head()
```

```python
# Select the rows with the total bill greater than 15
df[df["total_bill"] > 15]
```

```python
df["tip_percentage"] = np.round((df["tip"] / df["total_bill"]) * 100, decimals=2)
df.head()
```

```python
# Select the rows with multiple conditions
df[(df["tip_percentage"] > 20) & (df["sex"] == "Female")]
```

```python
# select the row using or condition
df[(df["tip_percentage"] > 15) | (df["tip"] > 15)]
```

```python
# Select the rows based on the options
options = [3,4]
df[df["size"].isin(options)]
```
